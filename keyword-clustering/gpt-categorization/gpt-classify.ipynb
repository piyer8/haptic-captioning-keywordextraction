```python
from openai import OpenAI
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from pydantic import BaseModel

extracted_answers = []
query_file = "queries.json"


dotenv_path = Path('../../global_environment.env')
load_dotenv(dotenv_path=dotenv_path)
openai_key = os.getenv('OPENAI_KEY')

client = OpenAI(
        api_key=openai_key
)

if(openai_key):
    print('Key loaded successfully')
else:
    print('API Key not found')
```

```python

def getRating(keywords):
    keywords_list = '\n'.join(keywords)
    
    response = client.beta.chat.completions.parse(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are an assistant that processes descriptions of haptic feedback and returns json objects"},
                {"role": "user", "content": "Classify the list of keywords and phrases into categories and return a json object containing the categories and a list of respective key words/phrases:\n"+keywords_list},
                ],
        )
    print(response.choices[0].message.content)
    return json.loads(response.choices[0].message.content) # response.choices[0].message.content

with open('./all_keywords.json', 'r') as f:
    words = json.load(f)

def save_json(filename, data):
    with open(filename, "w") as f:
        json.dump(data, f, indent=4)

# sensory_categories = getRating(words['sensory_keywords'])
# save_json('./sensory_categories.json', sensory_categories)
#emotional_categories = getRating(words['emotional_keywords'])
#save_json('./emotional_categories.json', emotional_categories)
associative_categories = getRating(words['associative_keywords'])
save_json('./associative_categories.json', associative_categories)
```
