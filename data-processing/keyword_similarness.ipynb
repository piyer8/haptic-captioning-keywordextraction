```python
api_endpoint = "http://api.conceptnet.io"

def calculate_similarity(keyword1, keyword2):
    url = f"{api_endpoint}/relatedness?node1=/c/en/{keyword1}&node2=/c/en/{keyword2}"
    response = requests.get(url)
    if response.status_code == 200:
        similarity_score = response.json()["value"]
        return similarity_score
    else:
        print(f"Error: Failed to calculate similarity for keywords {keyword1} and {keyword2}.")
        return None
```
