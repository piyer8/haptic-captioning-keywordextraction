```python
import json
import regex as re
import os

lexicon_path = './WorryWords/worrywords-v1.txt'

words = {}

with open(lexicon_path, 'r') as f:
    lines = f.read().splitlines()
    for line in lines[1:]:
        term,mean,ordinalClass,majorityLabel,majorityLabelRatio,numberAnnotations,setOfLabels = line.split('\t')
        words[term] = float(mean)

signals_path = '../../keyword-extraction/collated_descriptions.json'
words_scores_path = './word_worry_scores.json'
words_detailed_scores_path = './word_worry_scores_detailed.json'
existing_word_scores = {}

signal_scores_with_keywords = []

with open(signals_path, 'r') as f:
    signals = json.load(f)

def get_worry_score(signal):
    worry_score = 0
    count = 0
    previous = ''
    excluded_words = ['vibration', 'vibrations', 'is', 'was', 'like', 'a', 'sensation', 'your', 'too', 'made', 'as', 'on', 'just', 'when']
    descriptions = signal['emotional_descriptions']+signal['associative_descriptions']
    signal_words = {}
    for description in descriptions:
        wordlist = description.split(' ')
        for w in wordlist:
            word = w.lower()
            word = re.sub('[^A-Za-z0-9]+', '', word)
            if word not in excluded_words:
                if word in words:
                    score = words[word]
                    if score == 0:
                        continue
                    if previous == 'not':
                        worry_score -= score
                        signal_words['not ' + word] = -1 * score
                    else:
                        worry_score += score
                        signal_words[word] = score
                    count += 1
                    if word not in existing_word_scores:
                        existing_word_scores[word] = score
                    if previous == 'not' and 'not '+word not in existing_word_scores:
                        existing_word_scores['not '+word] = -1 * score
                else:
                     print('word not in lexicon: ' + word)
            previous = word
            
    signal_scores_with_keywords.append({'signal_index': signal['signal_index'], 'words': signal_words, 'worry_score': worry_score})
    if count != 0:
        return worry_score/count
    else:
        return 0

def main():
    worry_scores = []
    for signal in signals:
        worry_scores.append({'signal_index': signal['signal_index'], 'worry_score': get_worry_score(signal)})
    output_file = 'worry_scores.json'
    with open(output_file, 'w') as f:
        json.dump(worry_scores, f)

    with open(words_scores_path, 'w') as f:
        json.dump(existing_word_scores, f)

    with open(words_detailed_scores_path, 'w') as f:
        json.dump(signal_scores_with_keywords, f)

if __name__ == '__main__':
    main()
```
